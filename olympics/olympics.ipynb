```python
from openpyxl import load_workbook
from datetime import datetime, timedelta

wb = load_workbook("nanntonaku3333.xlsx")
src = wb["Sheet1"]

# 基準日（C2）
start_date = src.cell(row=2, column=3).value

if not isinstance(start_date, datetime):
    raise ValueError("C2 は日付である必要があります")

sheet_index = 2
current_date = start_date + timedelta(days=2)

while current_date.day <= 25:
    sheet_name = f"Sheet{sheet_index}"

    # 既にあれば削除
    if sheet_name in wb.sheetnames:
        wb.remove(wb[sheet_name])

    dst = wb.create_sheet(sheet_name)

    for r in range(1, src.max_row + 1):
        for c in range(1, src.max_column + 1):
            value = src.cell(row=r, column=c).value

            # ★ 2行目・C列以降だけ日付を+2日ずつ
            if r == 2 and c >= 3 and isinstance(value, datetime):
                value = value + timedelta(days=2 * (sheet_index - 1))

            dst.cell(row=r, column=c).value = value

    sheet_index += 1
    current_date += timedelta(days=2)

wb.save("nanntonaku3333.xlsx")
```
