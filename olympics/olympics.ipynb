```python
from openpyxl import load_workbook

wb = load_workbook("nanntonaku.xlsx")

src = wb['Sheet3']
dst = wb['sheet4']

# 名前行を走査（2行目〜）
for r in range(2, src.max_row + 1):
    name = src.cell(row=r, column=1).value

    # シート4で同じ名前を探す
    for dr in range(2, dst.max_row + 1):
        if dst.cell(row=dr, column=1).value == name:

            # 駅を横にコピー
            for c in range(2, src.max_column + 1):
                dst.cell(row=dr, column=c).value = src.cell(row=r, column=c).value

            break

wb.save("nanntonaku.xlsx")

```
