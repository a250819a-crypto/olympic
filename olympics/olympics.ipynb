```python
from openpyxl import load_workbook
from datetime import timedelta

wb = load_workbook("nanntonaku.xlsx")

src = wb["Sheet1"]
dst = wb["Sheet2"]

for r in range(1, src.max_row + 1):
    for c in range(1, src.max_column + 1):
        value = src.cell(row=r, column=c).value

        # B列（日付）だけ +2日
        if c == 2 and isinstance(value, (int, float)) is False:
            # Excelの日付（datetime型）の場合
            try:
                value = value + timedelta(days=2)
            except:
                pass

        dst.cell(row=r, column=c).value = value

wb.save("nanntonaku.xlsx")
```

```python
from openpyxl import load_workbook
from datetime import timedelta, datetime

wb = load_workbook("nanntonaku.xlsx")

src = wb["Sheet3"]
dst = wb["Sheet4"]

for r in range(1, src.max_row + 1):
    for c in range(1, src.max_column + 1):
        value = src.cell(row=r, column=c).value

        # B列が「日付」のときだけ +2日
        if c == 2 and isinstance(value, datetime):
            value = value + timedelta(days=2)

        dst.cell(row=r, column=c).value = value

wb.save("nanntonaku.xlsx")
```
