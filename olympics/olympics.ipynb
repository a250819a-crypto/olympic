```python
from openpyxl import load_workbook
from datetime import datetime, timedelta

wb = load_workbook("nanntonaku3333.xlsx")
src = wb["Sheet1"]

# Sheet2 がなければ作る
if "Sheet2" in wb.sheetnames:
    dst = wb["Sheet2"]
else:
    dst = wb.create_sheet("Sheet2")

for r in range(1, src.max_row + 1):
    for c in range(1, src.max_column + 1):
        value = src.cell(row=r, column=c).value

        # ★ 1行目かつB列以降 → 日付
        if r == 1 and c >= 2:
            if isinstance(value, datetime):
                value = value + timedelta(days=2)
            elif isinstance(value, str):
                try:
                    value = datetime.strptime(value, "%Y/%m/%d") + timedelta(days=2)
                except:
                    pass

        dst.cell(row=r, column=c).value = value

wb.save("nanntonaku3333.xlsx")
```
