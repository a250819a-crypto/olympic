```python
from openpyxl import load_workbook
from datetime import datetime, timedelta

wb = load_workbook("nanntonaku3333.xlsx")

src = wb["Sheet1"]

# 何枚作るか
num_sheets = 4  # Sheet2〜Sheet5

for i in range(1, num_sheets + 1):
    sheet_name = f"Sheet{i+1}"

    # すでにあればスキップ
    if sheet_name in wb.sheetnames:
        dst = wb[sheet_name]
    else:
        dst = wb.create_sheet(sheet_name)

    # ① 1行目（日付）を +2日ずつ増やす
    for c in range(1, src.max_column + 1):
        value = src.cell(row=1, column=c).value
        if isinstance(value, datetime):
            value = value + timedelta(days=2 * i)
        dst.cell(row=1, column=c).value = value

    # ② データ部分はそのままコピー
    for r in range(2, src.max_row + 1):
        for c in range(1, src.max_column + 1):
            dst.cell(row=r, column=c).value = src.cell(row=r, column=c).value

wb.save("nanntonaku3333.xlsx")
```
