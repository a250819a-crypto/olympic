```python
from openpyxl import load_workbook
from datetime import datetime, timedelta

wb = load_workbook("nanntonaku33333.xlsx")
src = wb["Sheet1"]

sheet_index = 2
add_days = 5

while True:
    sheet_name = f"Sheet{sheet_index}"

    if sheet_name in wb.sheetnames:
        wb.remove(wb[sheet_name])
    dst = wb.create_sheet(sheet_name)

    stop = False

    for r in range(1, src.max_row + 1):
        for c in range(1, src.max_column + 1):
            value = src.cell(row=r, column=c).value

            # ★ 2行目・C列以降（日付行）
            if r == 2 and c >= 3 and value is not None:
                try:
                    # 文字列 → datetime に変換
                    if isinstance(value, str):
                        value = datetime.strptime(value, "%Y/%m/%d")

                    value = value + timedelta(days=add_days * (sheet_index - 1))

                    # 25日を超えたら終了
                    if value.day > 25:
                        stop = True
                except:
                    pass

            dst.cell(row=r, column=c).value = value

    if stop:
        break

    sheet_index += 1

wb.save("sannba.xlsx")
```

```python
from openpyxl.utils import get_column_letter

for c in range(3, src.max_column + 1):
    col_letter = get_column_letter(c)
    dst.column_dimensions[col_letter].width = 12
```
