```python
from openpyxl import load_workbook
from datetime import datetime, timedelta

wb = load_workbook("nanntonaku3333.xlsx")

src = wb["Sheet1"]

# Sheet2 がなければ作成
if "Sheet2" in wb.sheetnames:
    dst = wb["Sheet2"]
else:
    dst = wb.create_sheet("Sheet2")

# ① 1行目（日付行）をコピーしつつ +2日
for c in range(1, src.max_column + 1):
    value = src.cell(row=1, column=c).value
    if isinstance(value, datetime):
        value = value + timedelta(days=2)
    dst.cell(row=1, column=c).value = value

# ② データ行をコピー
for r in range(2, src.max_row + 1):
    for c in range(1, src.max_column + 1):
        dst.cell(row=r, column=c).value = src.cell(row=r, column=c).value

wb.save("nanntonaku3333.xlsx")

```
